

### Data Cleaning 
    1. Dropping Duplicates and null values
    2. Dropping columns that are irrelevent to the analysis : id, budget_adj, imdb_id, homepage, tagline, revenue_adj,      keywords, overview.
    3. Dropping the column release_date as it contain many false values when compared to release_year column. 
    4. Dropping the null values from the director column
    5. Dropping rows with value 0 in budget and revenue as it will affect overall analysis and will also give false value to find profit made.

```python
# Dropping columns that are irrelevent to the analysis
df = df.drop(columns = ['id', 'budget_adj', 'imdb_id', 'homepage', 'tagline', 'revenue_adj', 'release_date','keywords','overview', 'runtime', 'popularity', 'production_companies'])
df.head(5)
```

```python
# dropping 0 values in budget column
df = df[df['budget'] != 0]

# checking changes
df[df['budget'] == 0]
```

```python
# dropping 0 values in revenue column
df = df[df['revenue'] != 0]

# checking changes
df[df['revenue'] == 0]
```

```python
# dropping nulls values in the dataset
df.dropna(inplace = True)

# checking changes
df.isnull().sum()
```

```python
df.head(5)
```

```python
df.shape
```

```python
# checking value in genres columns
df['genres'].unique()
```

```python
# genres column contains multiple values which can make analysis difficult
# splitting genres columns value with '|'
df = df.assign(genres = df['genres'].str.split('|')).explode('genres')

# checking changes
df.shape
```

```python
# checking if director column contains any multiple values
df['director'].unique()
```

```python
df = df.assign(director = df['director'].str.split('|')).explode('director')
df.shape
```

```python
df.duplicated().sum()
```

```python
# checking changes
df.head(5)
```

```python
# splitting cast column to get deeper analysis
df = df.assign(cast = df['cast'].str.split('|')).explode('cast')

# checking changes
df.shape
```

```python
# checking for duplicate value if any
df.duplicated().sum()
```

```python
# dropping duplicate values
df.drop_duplicates(inplace = True)

#checking changes
df.duplicated().sum()
```

```python
# saving clean dataset to another file
df.to_csv('cleaned-TMDB-dataset.csv', index_label = False)
```
